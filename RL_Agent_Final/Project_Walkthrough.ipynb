## 2. The Reward Function: Risk-Adjusted Logic

The goal of the agent is not just to win trades, but to **defend professional capital**. The reward function encodes high-level risk management rules.



*   **Trade Win**: `+8.0`
*   **Trade Loss**: `-7.2`
*   **End of Day (Profit)**: `+67.0` (Encourages stopping early with a win).
*   **End of Day (Loss)**: `-27.0`
*   **Daily Max Drawdown Hit**: `-21.0` (The "Fired" penalty).



## 3. The Best Model: Weighted Ensemble

We benchmarked PPO, DQN, QR-DQN, LSTM, and Transformer models. The ultimate winner was an ensemble where each agent provides a weighted vote.



### Ensemble Weights (Determined by Differential Evolution)
*   **DQN**: 1.32
*   **Transformer**: 1.22
*   **QR-DQN**: 1.05
*   **Recurrent**: 0.60
*   **PPO**: 0.37

**Decision Rule**: Vote Sum > **4.21**
